## 거주인구 데이터 전처리_월평균

```python
import pandas as pd

PATH = '../data/'
df = pd.read_csv(PATH + '2021_행정동별_거주인구_231015.csv')
```

```python
df
```

```python
dfs = df.groupby(['행정동코드', '연도'])[
    '연령대_I00', '연령대_0_거주인구수', '연령대_10_거주인구수', '연령대_20_거주인구수',
    '연령대_30_거주인구수', '연령대_40_거주인구수', '연령대_50_거주인구수',
    '연령대_60_거주인구수', '연령대_70_거주인구수', '연령대_80_거주인구수',
    '연령대_90_거주인구수'].sum() / 12
```

```python
dfs
```

```python
dfs.rename(columns = {'연령대_I00': '연령대_100'},inplace = True)
```

```python
dfs.reset_index(inplace = True)
```

```python
# 새로운 열 이름을 생성
new_column_names = {
    '연령대_100': '월평균_연령대_100',
    '연령대_0_거주인구수': '월평균_연령대_0_거주인구수',
    '연령대_10_거주인구수': '월평균_연령대_10_거주인구수',
    '연령대_20_거주인구수': '월평균_연령대_20_거주인구수',
    '연령대_30_거주인구수': '월평균_연령대_30_거주인구수',
    '연령대_40_거주인구수': '월평균_연령대_40_거주인구수',
    '연령대_50_거주인구수': '월평균_연령대_50_거주인구수',
    '연령대_60_거주인구수': '월평균_연령대_60_거주인구수',
    '연령대_70_거주인구수': '월평균_연령대_70_거주인구수',
    '연령대_80_거주인구수': '월평균_연령대_80_거주인구수',
    '연령대_90_거주인구수': '월평균_연령대_90_거주인구수'
}

# 열 이름을 변경
df.rename(columns=new_column_names, inplace=True)
```

```python
dfs.to_csv(PATH + '2021_행정동별_월평균_거주인구.csv',encoding = 'utf-8-sig', index = False)
```
