## 구별 데이터셋 전처리

```python
import pandas as pd

PATH = '../data/'
df = pd.read_csv(PATH + '2021_생활인구_행정동별 합계.csv')
dfs = pd.read_csv(PATH + '서울_행정동_법정동_맵핑데이터.csv')
df_1 = pd.read_csv(PATH + '2021_거주인구_편집.csv')
```

```python
df.info()
```

```python
df
```

```python
dfs
```

```python
dfc = dfs[['CTGG_NM','ADSTRD_CD']]
```

```python
cols = '자치구','행정동코드'
dfc.columns = cols
```

```python
dfc['행정동코드'] = dfc['행정동코드'].astype(str)
dfc['행정동코드'] = dfc['행정동코드'].str[:-2]
```

```python
dfc['행정동코드']
```

```python
df['행정동코드'] = df['행정동코드'].astype(str)
```

#### 행정동코드 대체(과거)

```python
df['행정동코드'] = df['행정동코드'].replace('11305590', '11305595')
df['행정동코드'] = df['행정동코드'].replace('11305600', '11305603')
df['행정동코드'] = df['행정동코드'].replace('11305606', '11305608')
df['행정동코드'] = df['행정동코드'].replace('11305610', '11305615')
df['행정동코드'] = df['행정동코드'].replace('11305620', '11305625')
df['행정동코드'] = df['행정동코드'].replace('11305630', '11305635')
df['행정동코드'] = df['행정동코드'].replace('11740520', '11740525')
```

```python
dfc = dfc.drop_duplicates(subset='행정동코드')
```

### 생활인구(행정동코드, 자치구) 병합, 구별 groupby

```python
merged_df = df.merge(dfc, on = '행정동코드', how = 'left')
```

```python
merged_df.isna().sum()
```

```python
grouped_df = merged_df.groupby(['자치구']).mean()
```

```python
grouped_df.reset_index(inplace = True)
grouped_df.rename(columns = {'행정동코드':'행정동_코드','총생활인구수':'구별_총생활인구수_평균','연령대0' : '구별_연령대0_생활인구_평균','연령대10' : '구별_연령대10_생활인구_평균','연령대20' : '구별_연령대20_생활인구_평균','연령대30' : '구별_연령대30_생활인구_평균',
                             '연령대40' : '구별_연령대40_생활인구_평균','연령대50' : '구별_연령대50_생활인구_평균','연령대60' : '구별_연령대60_생활인구_평균','연령대70이상' : '구별_연령대70_생활인구_평균'},inplace = True)
```

```python
grouped_df
```

```python
grouped_df.to_csv(PATH + '2021_생활인구_구별_집계.csv', index = False, encoding = 'utf-8-sig')
```
