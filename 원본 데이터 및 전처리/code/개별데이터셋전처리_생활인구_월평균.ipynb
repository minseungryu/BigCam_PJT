```python
import pandas as pd
PATH = '/content/drive/MyDrive/Colab Notebooks/공모전/data/'
df = pd.read_csv(PATH + 'LOCAL_PEOPLE_DONG_2021(merged).csv')
```

#### 2. 전처리

- 년도/월 추출

```python
df['기준일ID'] = df['기준일ID'].astype(str)

# '일자' 열에서 연도와 월 추출
df['년도'] = df['기준일ID'].str[:4]
df['월'] = df['기준일ID'].str[4:6]
```

- 컬럼 순서 변경

```python
# 열 순서 변경
df = df[['년도', '월'] + [col for col in df.columns if col not in ['년도', '월']]]
df
```

- 컬럼 삭제

```python
del_col = ['기준일ID','시간대구분']
df.drop(columns = del_col, inplace = True)
```

- 연령대별 컬럼 합산

```python
df['연령대0'] = df['남자0세부터9세생활인구수'] + df['여자0세부터9세생활인구수']
```

```python
df['연령대70이상'] =  df['남자70세이상생활인구수'] + df['여자70세이상생활인구수']
```

```python
for i in range(1, 7):
    i *= 10
    age_column = f'{i}세부터{i+4}세'

    # 남자와 여자의 해당 연령대 합계 계산
    df[f'연령대{i}'] = df[f'남자{age_column}생활인구수'] + df[f'여자{age_column}생활인구수'] + df[f'남자{i+5}세부터{i+9}세생활인구수'] + df[f'여자{i+5}세부터{i+9}세생활인구수']
```

- 필요없는 컬럼 삭제

```python
for i in range(1, 7):
    i *= 10
    age_column = f'{i}세부터{i+4}세'
    df.drop([f'남자{age_column}생활인구수', f'여자{age_column}생활인구수', f'남자{i+5}세부터{i+9}세생활인구수', f'여자{i+5}세부터{i+9}세생활인구수'], axis=1, inplace=True)
```

```python
df.drop(columns = ['남자0세부터9세생활인구수','여자0세부터9세생활인구수','남자70세이상생활인구수','여자70세이상생활인구수'])
```

```python
# 컬럼 순서 변경
new_column_order = ['년도','월','행정동코드','총생활인구수','연령대0','연령대10', '연령대20', '연령대30', '연령대40', '연령대50', '연령대60', '연령대70이상']
df = df[new_column_order]
```

```python
final_df = df.groupby(['년도','월','행정동코드']).sum()
```
