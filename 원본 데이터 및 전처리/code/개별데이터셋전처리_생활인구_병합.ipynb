### 1. 데이터 병합(21년)

```python
import pandas as pd

# 12개의 CSV 파일의 파일 이름을 리스트에 저장
csv_files = ['rawdata/LOCAL_PEOPLE_DONG_202101.csv', 'rawdata/LOCAL_PEOPLE_DONG_202102.csv',
             'rawdata/LOCAL_PEOPLE_DONG_202103.csv', 'rawdata/LOCAL_PEOPLE_DONG_202104.csv',
             'rawdata/LOCAL_PEOPLE_DONG_202105.csv', 'rawdata/LOCAL_PEOPLE_DONG_202106.csv',
             'rawdata/LOCAL_PEOPLE_DONG_202107.csv', 'rawdata/LOCAL_PEOPLE_DONG_202108.csv',
             'rawdata/LOCAL_PEOPLE_DONG_202109.csv', 'rawdata/LOCAL_PEOPLE_DONG_202110.csv',
             'rawdata/LOCAL_PEOPLE_DONG_202111.csv', 'rawdata/LOCAL_PEOPLE_DONG_202112.csv']

# 결과를 저장할 빈 데이터프레임 생성
combined_df = pd.DataFrame()

# 각 파일에 대해 작업 수행
for file in csv_files:
    # CSV 파일 읽기
    PATH = '../data/'
    df = pd.read_csv(PATH + file)

    # 인덱스 재설정
    df.reset_index(inplace=True)

    # 컬럼 이름 수정
    cols = df.columns.tolist()
    cols.pop(0)  # 첫 번째 컬럼 제거
    df = df.drop(df.columns[-1], axis=1)  # 마지막 컬럼 제거
    df.columns = cols

    # 데이터를 기존 데이터프레임에 추가
    combined_df = pd.concat([combined_df, df])

# 결과 데이터프레임 출력
combined_df
```

```python
PATH = '/content/drive/MyDrive/Colab Notebooks/공모전/data/'
combined_df.to_csv(PATH + 'LOCAL_PEOPLE_DONG_2021(merged).csv',index = False, encoding = 'utf-8-sig')
```
