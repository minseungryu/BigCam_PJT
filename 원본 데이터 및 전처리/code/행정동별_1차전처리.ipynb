### 1.전처리

#### df1 : 2015 행정동 소득수준
- unique 한 행정동 코드만 인덱스로 두고 연령대별 소득구간 컬럼을 많아서 행정동 코드당 하나의 행이 만들어지게 해야함

```python
df1
```

```python
df1_encoded = pd.get_dummies(df1, columns=['연령대'])
df1_encoded.head()
```

```python
# '소득구간' 값이 1인 경우 해당 연령대 컬럼에 소득구간 값을 넣기(소득구간 * 1)
age_columns = ['연령대_20대', '연령대_30대', '연령대_40대', '연령대_50대', '연령대_60세이상']
for age_column in age_columns:
    df1_encoded[age_column] = df1_encoded[age_column] * df1_encoded['소득구간']

df1_encoded.drop(columns = ['소득구간'], inplace = True)
df1_encoded.head()
```

```python
df_1 = df1_encoded.groupby('행정동코드').sum()
df_1.reset_index(inplace = True)
```

```python
cols = ['행정동코드','연령대_20대_소득구간','연령대_30대_소득구간','연령대_40대_소득구간','연령대_50대_소득구간','연령대_60세이상_소득구간']
df_1.columns = cols
df_1.head()
```

```python
df_1.to_csv(PATH + '2015_행정동별_연령대별_소득구간_인코딩_231015.csv', index = False, encoding = 'utf-8-sig')
```

```python
df6.drop(columns = ['성별코드'],inplace = True)
df6 = df6.groupby(['행정동코드','행정동','자치구','연도','월','연령대코드']).sum().reset_index()
df6_encoded = pd.get_dummies(df6, columns=['연령대코드'], prefix='연령대')
```

```python
df6 = df6.groupby(['행정동코드','행정동','자치구','연도','월','연령대코드']).sum().reset_index()
```

```python
df6
```

```python
df6_encoded = pd.get_dummies(df6, columns=['연령대코드'], prefix='연령대')
df6_encoded
```

```python
cols = df6_encoded.columns.to_list()
cols
```

```python
# '연령대_' 값이 1인 경우 해당 연령대 컬럼에 소득구간 값을 넣기(거주인구수 * 1)
age_columns = cols[6:]
for age_column in age_columns:
    df6_encoded[age_column] = df6_encoded[age_column] * df6_encoded['거주인구수']

df6_encoded.drop(columns = ['거주인구수'], inplace = True)
df6_encoded.head()
```

### 2.merge

2021 생활인구와 거주인구 데이터 행정동코드로 merge

```python
df = pd.read_csv(PATH +'2021_행정동별_거주인구_231015.csv') ## 월별로 구분되어 있음
data = pd.read_csv(PATH +'2021_생활인구_행정동별_합계.csv') ## 월별 합산되어있음
df_1 = pd.read_csv(PATH +'2021_행정동별_월별_생활인구.csv') ## 월별 구분
```

```python
df_1.rename(columns = {'년도':'연도','연령대0':'연령대0_생활인구','연령대10':'연령대10_생활인구','연령대20':'연령대20_생활인구','연령대30':'연령대30_생활인구','연령대40':'연령대40_생활인구','연령대50':'연령대50_생활인구',
            '연령대60':'연령대60_생활인구','연령대70이상':'연령대70이상_생활인구'},inplace = True)
```

```python
merged_df = df.merge(df_1, on = ['연도','월','행정동코드'], how = 'outer')
```
